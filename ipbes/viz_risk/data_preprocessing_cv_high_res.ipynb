```python
import geopandas as gpd
import fiona
import pandas as pd
```

```python
save_dir = "./data/preprocessed_data/updated_data3/"
```

```python
def poly_to_lat_long(poly):

    lats = list(poly.exterior.coords.xy[1])
    longs = list(poly.exterior.coords.xy[0])
    return list(zip(lats, longs))

def point_to_lat_long(point):
    return (point.y, point.x)
```

```python
gpkg_data_dir = "./data/gpkg_data/"

cv_highres_name = "cv_coastal_points_output_md5_69641307c3c7b4c7d23faa8637e30f83.gpkg"
fiona.listlayers(gpkg_data_dir + cv_highres_name)
```

```python
cv_highres_data = gpd.read_file(gpkg_data_dir + cv_highres_name, layer='CV_outputs')
```

```python
cv_highres_data.shape
```

```python
cv_highres_data = cv_highres_data.sample(30000)
```

```python
cv_highres_data.columns
```

```python
remainder = cv_highres_data[['Rt_cur', 'Rt_ssp1', 'Rt_ssp3', 'Rt_ssp5', 'NCP_cur', 'NCP_ssp1', 'NCP_ssp3', 'NCP_ssp5', 'geometry']]
```

```python
remainder.geometry.head()
remainder['coords'] = remainder['geometry'].apply(lambda x: point_to_lat_long(x))
remainder
```

```python
remainder[['lat', 'lng']] = pd.DataFrame(remainder['coords'].tolist(), index=remainder.index)
remainder = remainder.drop(columns=['geometry', 'coords'])
remainder.head()
```

```python
remainder.columns = ['UN_c', 'UN_1', 'UN_3', 'UN_5', 'NC_c', 'NC_1', 'NC_3', 'NC_5', 'lat', 'lng']
remainder.sample(5)
```

```python
remainder.reset_index(drop=True, inplace=True)
```

```python
remainder.to_csv(save_dir + "cv_high_res_table_preprocessed.csv", index=False)
```
